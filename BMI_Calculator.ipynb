# BMI Calculator
This notebook calculates the Body Mass Index (BMI) using your height and weight.

It uses interactive widgets via `ipywidgets`.

```python
import ipywidgets as widgets
from IPython.display import display, Markdown

# Widgets for input
height = widgets.FloatSlider(value=1.70, min=1.0, max=2.2, step=0.01, description='Height (m):')
weight = widgets.FloatSlider(value=70, min=30, max=200, step=1, description='Weight (kg):')
button = widgets.Button(description="Calculate BMI")

output = widgets.Output()

def calculate_bmi(b):
    with output:
        output.clear_output()
        bmi = weight.value / (height.value ** 2)
        if bmi < 18.5:
            status = 'Underweight'
        elif bmi < 25:
            status = 'Normal weight'
        elif bmi < 30:
            status = 'Overweight'
        else:
            status = 'Obese'
        display(Markdown(f"**Your BMI is:** {bmi:.2f} ({status})"))

button.on_click(calculate_bmi)

display(height, weight, button, output)
```
